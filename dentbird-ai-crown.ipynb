```python
import pandas as pd
import random
import os

manifest = pd.read_csv("/kaggle/working/manifest.csv")
manifest = manifest[manifest["is_trainable"] == True].copy()

case_ids = manifest["case_id"].tolist()
random.seed(42)
random.shuffle(case_ids)

n = len(case_ids)
n_train = int(0.7 * n)
n_val = int(0.15 * n)

train_ids = case_ids[:n_train]
val_ids = case_ids[n_train:n_train + n_val]
test_ids = case_ids[n_train + n_val:]

splits_dir = "/kaggle/working/splits"
os.makedirs(splits_dir, exist_ok=True)

for name, ids in [("train.txt", train_ids), ("val.txt", val_ids), ("test.txt", test_ids)]:
    with open(os.path.join(splits_dir, name), "w", encoding="utf-8") as f:
        f.write("\n".join(ids))

print("train:", len(train_ids))
print("val:", len(val_ids))
print("test:", len(test_ids))
print("saved to:", splits_dir)
```

```python
baseline_code = r'''
from __future__ import annotations

import torch
import torch.nn as nn


class PointEncoder(nn.Module):
    def __init__(self, in_dim: int = 3, hidden: int = 128, out_dim: int = 256):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(in_dim, hidden),
            nn.ReLU(),
            nn.Linear(hidden, hidden),
            nn.ReLU(),
            nn.Linear(hidden, out_dim),
        )

    def forward(self, x: torch.Tensor) -> torch.Tensor:
        x = torch.nan_to_num(x, nan=0.0, posinf=1e3, neginf=-1e3)
        feat = self.net(x)
        feat = torch.nan_to_num(feat, nan=0.0, posinf=1e3, neginf=-1e3)
        return feat.max(dim=1).values


class CrownRefiner(nn.Module):
    def __init__(self, latent_dim: int = 256, num_points: int = 4096, use_bite: bool = True):
        super().__init__()
        self.num_points = num_points
        self.use_bite = use_bite

        self.upper_enc = PointEncoder(out_dim=latent_dim)
        self.lower_enc = PointEncoder(out_dim=latent_dim)
        self.tooth_enc = PointEncoder(out_dim=latent_dim)
        self.bite_enc = PointEncoder(out_dim=latent_dim)

        fusion_in = latent_dim * (4 if use_bite else 3)
        self.fusion = nn.Sequential(
            nn.Linear(fusion_in, latent_dim),
            nn.ReLU(),
            nn.Linear(latent_dim, latent_dim),
            nn.ReLU(),
        )

        self.decoder = nn.Sequential(
            nn.Linear(latent_dim + 3, latent_dim),
            nn.ReLU(),
            nn.Linear(latent_dim, latent_dim),
            nn.ReLU(),
            nn.Linear(latent_dim, 3),
        )

    def forward(
        self,
        upper: torch.Tensor,
        lower: torch.Tensor,
        tooth_model: torch.Tensor,
        bite: torch.Tensor | None = None
    ) -> torch.Tensor:
        upper = torch.nan_to_num(upper, nan=0.0, posinf=1e3, neginf=-1e3)
        lower = torch.nan_to_num(lower, nan=0.0, posinf=1e3, neginf=-1e3)
        tooth_model = torch.nan_to_num(tooth_model, nan=0.0, posinf=1e3, neginf=-1e3)

        f_upper = self.upper_enc(upper)
        f_lower = self.lower_enc(lower)
        f_tooth = self.tooth_enc(tooth_model)

        feats = [f_upper, f_lower, f_tooth]

        if self.use_bite:
            bite = bite if bite is not None else torch.zeros_like(tooth_model)
            bite = torch.nan_to_num(bite, nan=0.0, posinf=1e3, neginf=-1e3)
            feats.append(self.bite_enc(bite))

        fused = self.fusion(torch.cat(feats, dim=-1))
        fused = torch.nan_to_num(fused, nan=0.0, posinf=1e3, neginf=-1e3)

        repeated = fused.unsqueeze(1).repeat(1, tooth_model.shape[1], 1)
        dec_in = torch.cat([repeated, tooth_model], dim=-1)

        raw_offsets = self.decoder(dec_in)
        raw_offsets = torch.nan_to_num(raw_offsets, nan=0.0, posinf=1e3, neginf=-1e3)

        # bounded deformation
        offsets = torch.tanh(raw_offsets) * 0.05
        pred = tooth_model + offsets
        pred = torch.nan_to_num(pred, nan=0.0, posinf=1e3, neginf=-1e3)
        return pred
'''
with open("/kaggle/working/dentbird_project/dentbird/models/baseline_refiner.py", "w", encoding="utf-8") as f:
    f.write(baseline_code)

print("Patched baseline_refiner.py ✅")
```

```python
losses_code = r'''
from __future__ import annotations

import torch
import torch.nn.functional as F


def _clean(x: torch.Tensor) -> torch.Tensor:
    return torch.nan_to_num(x, nan=0.0, posinf=1e4, neginf=-1e4)


def pairwise_sqdist(x: torch.Tensor, y: torch.Tensor) -> torch.Tensor:
    x = _clean(x)
    y = _clean(y)
    x2 = (x ** 2).sum(dim=-1, keepdim=True)
    y2 = (y ** 2).sum(dim=-1).unsqueeze(1)
    xy = x @ y.transpose(1, 2)
    dist = x2 + y2 - 2 * xy
    return dist.clamp_min(0.0)


def chamfer_distance(x: torch.Tensor, y: torch.Tensor) -> torch.Tensor:
    dist = pairwise_sqdist(x, y)
    x_to_y = dist.min(dim=2).values.mean(dim=1)
    y_to_x = dist.min(dim=1).values.mean(dim=1)
    out = (x_to_y + y_to_x).mean()
    return _clean(out)


def approx_hausdorff_distance(x: torch.Tensor, y: torch.Tensor, percentile: float = 0.95) -> torch.Tensor:
    dist = pairwise_sqdist(x, y).sqrt()
    x_to_y = dist.min(dim=2).values
    y_to_x = dist.min(dim=1).values
    qx = torch.quantile(x_to_y, percentile, dim=1)
    qy = torch.quantile(y_to_x, percentile, dim=1)
    out = torch.maximum(qx, qy).mean()
    return _clean(out)


def gather_margin_points(pred: torch.Tensor, margin_idx: torch.Tensor) -> torch.Tensor:
    idx = margin_idx.unsqueeze(-1).expand(-1, -1, pred.shape[-1])
    gathered = torch.gather(pred, dim=1, index=idx)
    return _clean(gathered)


def margin_loss(pred: torch.Tensor, design_margin: torch.Tensor, margin_idx: torch.Tensor) -> torch.Tensor:
    pred_margin = gather_margin_points(pred, margin_idx)
    return chamfer_distance(pred_margin, design_margin)


def _weighted_point_mse(pred: torch.Tensor, ref: torch.Tensor, weight: torch.Tensor) -> torch.Tensor:
    pred = _clean(pred)
    ref = _clean(ref)
    weight = _clean(weight)
    sq = ((pred - ref) ** 2).sum(dim=-1)
    w = weight.float()
    denom = w.sum().clamp_min(1.0)
    out = (sq * w).sum() / denom
    return _clean(out)


def region_preservation_loss(
    pred: torch.Tensor,
    prior: torch.Tensor,
    margin_mask: torch.Tensor,
    axial_mask: torch.Tensor,
    occlusal_mask: torch.Tensor,
    w_margin: float = 4.0,
    w_axial: float = 1.5,
    w_occlusal: float = 0.25,
) -> torch.Tensor:
    l_margin = _weighted_point_mse(pred, prior, margin_mask)
    l_axial = _weighted_point_mse(pred, prior, axial_mask)
    l_occlusal = _weighted_point_mse(pred, prior, occlusal_mask)
    return _clean(w_margin * l_margin + w_axial * l_axial + w_occlusal * l_occlusal)


def masked_chamfer_distance(
    pred: torch.Tensor,
    target: torch.Tensor,
    pred_mask: torch.Tensor,
    target_mask: torch.Tensor | None = None,
) -> torch.Tensor:
    pred = _clean(pred)
    target = _clean(target)
    pred_mask = _clean(pred_mask)
    if target_mask is None:
        target_mask = pred_mask
    target_mask = _clean(target_mask)

    dist = pairwise_sqdist(pred, target)

    pred_to_target = dist.min(dim=2).values
    wp = pred_mask.float()
    pred_term = (pred_to_target * wp).sum() / wp.sum().clamp_min(1.0)

    target_to_pred = dist.min(dim=1).values
    wt = target_mask.float()
    target_term = (target_to_pred * wt).sum() / wt.sum().clamp_min(1.0)

    return _clean(pred_term + target_term)


def region_target_loss(
    pred: torch.Tensor,
    target: torch.Tensor,
    margin_mask: torch.Tensor,
    axial_mask: torch.Tensor,
    occlusal_mask: torch.Tensor,
    design_axial_mask: torch.Tensor | None = None,
    design_occlusal_mask: torch.Tensor | None = None,
    w_margin: float = 2.0,
    w_axial: float = 1.0,
    w_occlusal: float = 2.0,
) -> torch.Tensor:
    if design_axial_mask is None:
        design_axial_mask = axial_mask
    if design_occlusal_mask is None:
        design_occlusal_mask = occlusal_mask

    l_margin = masked_chamfer_distance(pred, target, margin_mask, margin_mask)
    l_axial = masked_chamfer_distance(pred, target, axial_mask, design_axial_mask)
    l_occlusal = masked_chamfer_distance(pred, target, occlusal_mask, design_occlusal_mask)

    return _clean(w_margin * l_margin + w_axial * l_axial + w_occlusal * l_occlusal)


def occlusal_clearance_band_penalty(
    pred: torch.Tensor,
    lower: torch.Tensor,
    occlusal_mask: torch.Tensor,
    min_clearance_norm: torch.Tensor,
    max_clearance_norm: torch.Tensor,
) -> torch.Tensor:
    pred = _clean(pred)
    lower = _clean(lower)
    occlusal_mask = _clean(occlusal_mask)
    min_clearance_norm = _clean(min_clearance_norm)
    max_clearance_norm = _clean(max_clearance_norm)

    dist = pairwise_sqdist(pred, lower).sqrt()
    nearest = dist.min(dim=2).values

    min_c = min_clearance_norm[:, None]
    max_c = max_clearance_norm[:, None]

    low_pen = F.relu(min_c - nearest).pow(2)
    high_pen = F.relu(nearest - max_c).pow(2)

    w = occlusal_mask.float()
    denom = w.sum().clamp_min(1.0)
    out = ((low_pen + high_pen) * w).sum() / denom
    return _clean(out)


def composite_crown_loss(
    pred: torch.Tensor,
    target: torch.Tensor,
    lower: torch.Tensor,
    design_margin: torch.Tensor,
    margin_idx: torch.Tensor,
    scale: torch.Tensor,
    tooth_model: torch.Tensor | None = None,
    margin_mask: torch.Tensor | None = None,
    axial_mask: torch.Tensor | None = None,
    occlusal_mask: torch.Tensor | None = None,
    design_axial_mask: torch.Tensor | None = None,
    design_occlusal_mask: torch.Tensor | None = None,
    min_clearance_mm: float = 0.5,
    max_clearance_mm: float = 2.0,
    w_chamfer: float = 1.0,
    w_margin: float = 3.0,
    w_hausdorff: float = 0.25,
    w_occlusion: float = 0.5,
    w_prior: float = 0.01,
    w_region_target: float = 0.6,
) -> tuple[torch.Tensor, dict[str, float]]:
    pred = _clean(pred)
    target = _clean(target)
    lower = _clean(lower)
    design_margin = _clean(design_margin)
    scale = _clean(scale)

    cham = chamfer_distance(pred, target)
    marg = margin_loss(pred, design_margin, margin_idx)
    haus = approx_hausdorff_distance(pred, target, percentile=0.95)

    if occlusal_mask is None:
        occlusal_mask = torch.zeros(pred.shape[:2], device=pred.device, dtype=pred.dtype)

    min_clearance_norm = (torch.tensor(min_clearance_mm, device=pred.device) / scale.clamp_min(1e-6)).float()
    max_clearance_norm = (torch.tensor(max_clearance_mm, device=pred.device) / scale.clamp_min(1e-6)).float()

    occ = occlusal_clearance_band_penalty(
        pred, lower, occlusal_mask, min_clearance_norm, max_clearance_norm
    )

    prior_term = torch.tensor(0.0, device=pred.device)
    region_term = torch.tensor(0.0, device=pred.device)

    if margin_mask is not None and axial_mask is not None and occlusal_mask is not None:
        region_term = region_target_loss(
            pred, target,
            margin_mask, axial_mask, occlusal_mask,
            design_axial_mask=design_axial_mask,
            design_occlusal_mask=design_occlusal_mask,
        )

    if w_prior > 0.0 and tooth_model is not None and margin_mask is not None and axial_mask is not None and occlusal_mask is not None:
        tooth_model = _clean(tooth_model)
        prior_term = region_preservation_loss(
            pred, tooth_model, margin_mask, axial_mask, occlusal_mask
        )

    total = (
        w_chamfer * cham
        + w_margin * marg
        + w_hausdorff * haus
        + w_occlusion * occ
        + w_prior * prior_term
        + w_region_target * region_term
    )
    total = _clean(total)

    metrics = {
        "loss_total": float(total.item()),
        "loss_chamfer": float(_clean(cham).item()),
        "loss_margin": float(_clean(marg).item()),
        "loss_hausdorff95": float(_clean(haus).item()),
        "loss_occlusion": float(_clean(occ).item()),
        "loss_prior": float(_clean(prior_term).item()),
        "loss_region_target": float(_clean(region_term).item()),
    }
    return total, metrics
'''
with open("/kaggle/working/dentbird_project/dentbird/models/losses.py", "w", encoding="utf-8") as f:
    f.write(losses_code)

print("Patched losses.py ✅")
```

```python
train_code = r'''
from __future__ import annotations

import argparse
import csv
from pathlib import Path

import torch
from torch.utils.data import DataLoader
from tqdm import tqdm

from dentbird.data.dataset import DentBirdNPZDataset
from dentbird.models.baseline_refiner import CrownRefiner
from dentbird.models.losses import composite_crown_loss


def _tensor_finite(x: torch.Tensor) -> bool:
    return torch.isfinite(x).all().item()


def _batch_is_finite(batch) -> bool:
    for k, v in batch.items():
        if isinstance(v, torch.Tensor):
            if not _tensor_finite(v):
                print(f"[WARN] Non-finite tensor in batch key: {k}")
                return False
    return True


def run_epoch(model, loader, optimizer, device, train: bool, min_clearance_mm: float, max_clearance_mm: float):
    model.train(train)
    total = 0.0
    count = 0
    skipped = 0

    agg = {
        "loss_chamfer": 0.0,
        "loss_margin": 0.0,
        "loss_hausdorff95": 0.0,
        "loss_occlusion": 0.0,
        "loss_prior": 0.0,
        "loss_region_target": 0.0,
    }

    for batch in tqdm(loader, leave=False):
        if not _batch_is_finite(batch):
            skipped += 1
            continue

        upper = batch["upper_local"].to(device)
        lower = batch["lower_local"].to(device)
        tooth = batch["tooth_model"].to(device)
        bite = batch["bite"].to(device)
        target = batch["design"].to(device)
        design_margin = batch["design_margin"].to(device)
        margin_idx = batch["margin_idx"].to(device)
        scale = batch["scale"].to(device)

        margin_mask = batch["margin_mask"].to(device)
        axial_mask = batch["axial_mask"].to(device)
        occlusal_mask = batch["occlusal_mask"].to(device)

        design_axial_mask = batch["design_axial_mask"].to(device)
        design_occlusal_mask = batch["design_occlusal_mask"].to(device)

        upper = torch.nan_to_num(upper, nan=0.0, posinf=1e3, neginf=-1e3)
        lower = torch.nan_to_num(lower, nan=0.0, posinf=1e3, neginf=-1e3)
        tooth = torch.nan_to_num(tooth, nan=0.0, posinf=1e3, neginf=-1e3)
        bite = torch.nan_to_num(bite, nan=0.0, posinf=1e3, neginf=-1e3)
        target = torch.nan_to_num(target, nan=0.0, posinf=1e3, neginf=-1e3)
        design_margin = torch.nan_to_num(design_margin, nan=0.0, posinf=1e3, neginf=-1e3)
        scale = torch.nan_to_num(scale, nan=1.0, posinf=1.0, neginf=1.0)

        pred = model(upper, lower, tooth, bite)

        if not torch.isfinite(pred).all():
            print("[WARN] Non-finite prediction, batch skipped")
            skipped += 1
            continue

        loss, metrics = composite_crown_loss(
            pred=pred,
            target=target,
            lower=lower,
            design_margin=design_margin,
            margin_idx=margin_idx,
            scale=scale,
            tooth_model=tooth,
            margin_mask=margin_mask,
            axial_mask=axial_mask,
            occlusal_mask=occlusal_mask,
            design_axial_mask=design_axial_mask,
            design_occlusal_mask=design_occlusal_mask,
            min_clearance_mm=min_clearance_mm,
            max_clearance_mm=max_clearance_mm,
        )

        if not torch.isfinite(loss):
            print("[WARN] Non-finite loss, batch skipped")
            skipped += 1
            continue

        if train:
            optimizer.zero_grad(set_to_none=True)
            loss.backward()
            torch.nn.utils.clip_grad_norm_(model.parameters(), 0.5)

            bad_grad = False
            for p in model.parameters():
                if p.grad is not None and not torch.isfinite(p.grad).all():
                    bad_grad = True
                    break

            if bad_grad:
                print("[WARN] Non-finite gradient, batch skipped")
                optimizer.zero_grad(set_to_none=True)
                skipped += 1
                continue

            optimizer.step()

            bad_param = False
            for p in model.parameters():
                if not torch.isfinite(p).all():
                    bad_param = True
                    break

            if bad_param:
                print("[WARN] Non-finite parameter after step, stopping epoch early")
                optimizer.zero_grad(set_to_none=True)
                break

        total += float(loss.item())
        count += 1
        for k in agg:
            agg[k] += metrics[k]

    out = {k: (v / max(count, 1)) for k, v in agg.items()}
    out["loss_total"] = total / max(count, 1)
    out["num_used_batches"] = count
    out["num_skipped_batches"] = skipped
    return out


def main() -> None:
    parser = argparse.ArgumentParser()
    parser.add_argument("--processed-dir", required=True)
    parser.add_argument("--splits-dir", required=True)
    parser.add_argument("--epochs", type=int, default=20)
    parser.add_argument("--batch-size", type=int, default=4)
    parser.add_argument("--lr", type=float, default=1e-4)
    parser.add_argument("--num-workers", type=int, default=0)
    parser.add_argument("--output-dir", default="outputs/checkpoints")
    parser.add_argument("--min-clearance-mm", type=float, default=0.8)
    parser.add_argument("--max-clearance-mm", type=float, default=1.5)
    args = parser.parse_args()

    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    out_dir = Path(args.output_dir)
    out_dir.mkdir(parents=True, exist_ok=True)
    metrics_path = out_dir / "metrics.csv"

    train_ds = DentBirdNPZDataset(args.processed_dir, Path(args.splits_dir) / "train.txt")
    val_ds = DentBirdNPZDataset(args.processed_dir, Path(args.splits_dir) / "val.txt")

    train_loader = DataLoader(train_ds, batch_size=args.batch_size, shuffle=True, num_workers=args.num_workers)
    val_loader = DataLoader(val_ds, batch_size=args.batch_size, shuffle=False, num_workers=args.num_workers)

    model = CrownRefiner().to(device)
    optimizer = torch.optim.Adam(model.parameters(), lr=args.lr)

    best_val = float("inf")
    with metrics_path.open("w", newline="", encoding="utf-8") as f:
        writer = csv.DictWriter(
            f,
            fieldnames=[
                "epoch",
                "train_loss_total", "train_loss_chamfer", "train_loss_margin",
                "train_loss_hausdorff95", "train_loss_occlusion", "train_loss_prior",
                "train_loss_region_target", "train_num_used_batches", "train_num_skipped_batches",
                "val_loss_total", "val_loss_chamfer", "val_loss_margin",
                "val_loss_hausdorff95", "val_loss_occlusion", "val_loss_prior",
                "val_loss_region_target", "val_num_used_batches", "val_num_skipped_batches",
            ],
        )
        writer.writeheader()

        for epoch in range(1, args.epochs + 1):
            train_m = run_epoch(model, train_loader, optimizer, device, True, args.min_clearance_mm, args.max_clearance_mm)
            with torch.no_grad():
                val_m = run_epoch(model, val_loader, optimizer, device, False, args.min_clearance_mm, args.max_clearance_mm)

            print(
                f"Epoch {epoch:03d} | "
                f"train={train_m['loss_total']:.6f} "
                f"(ch={train_m['loss_chamfer']:.4f}, mg={train_m['loss_margin']:.4f}, "
                f"rg={train_m['loss_region_target']:.4f}, pr={train_m['loss_prior']:.4f}, "
                f"oc={train_m['loss_occlusion']:.4f}, skip={train_m['num_skipped_batches']}) | "
                f"val={val_m['loss_total']:.6f} "
                f"(ch={val_m['loss_chamfer']:.4f}, mg={val_m['loss_margin']:.4f}, "
                f"rg={val_m['loss_region_target']:.4f}, pr={val_m['loss_prior']:.4f}, "
                f"oc={val_m['loss_occlusion']:.4f}, skip={val_m['num_skipped_batches']})"
            )

            row = {
                "epoch": epoch,
                **{f"train_{k}": v for k, v in train_m.items()},
                **{f"val_{k}": v for k, v in val_m.items()},
            }
            writer.writerow(row)
            f.flush()

            ckpt = {
                "epoch": epoch,
                "model_state": model.state_dict(),
                "optimizer_state": optimizer.state_dict(),
                "val_loss": val_m["loss_total"],
                "config": vars(args),
            }
            torch.save(ckpt, out_dir / "last.pt")
            if val_m["loss_total"] < best_val:
                best_val = val_m["loss_total"]
                torch.save(ckpt, out_dir / "best.pt")


if __name__ == "__main__":
    main()
'''
with open("/kaggle/working/dentbird_project/train.py", "w", encoding="utf-8") as f:
    f.write(train_code)

print("Patched train.py ✅")
```
